# Makeup Sales

## Telecharger data

```python
import pandas as pd

df = pd.read_csv('makeup_sales_dataset_2025.csv')
df.columns
```

```python
df.loc[(df['Brand']=="MAC") & (df['Product_Type']=="Eyeliner")]
```

## Statistiques

### Moyenne de revenue des produits de Dior dans tous les pays

```python
df.loc[df['Brand'] == "Dior"].groupby(['Product_Type'])['Revenue_USD'].mean()
```

### Liste des produits les plus rentables

```python
df.groupby(['Product_Type'])['Revenue_USD'].sum().sort_values(ascending=False)
```

### Liste des produits les plus vendus

```python
df.groupby(['Product_Type'])['Units_Sold'].sum().sort_values(ascending=False)
```

### Liste des pays qui gagnent plus

```python
df.groupby(['Country'])['Revenue_USD'].sum().sort_values(ascending=True)
```

## Top brand

```python
df.groupby(['Brand'])['Revenue_USD'].sum().sort_values(ascending=False)
```

## Changer le data

### Changer le prix de Eyeliner de Eyeliner de MAC pour sale event dans tous les pays

```python
filtre = (df['Brand'] == "MAC") & (df['Product_Type'] == "Eyeliner")

df.loc[filtre, 'Price_USD'] = 15.5
df.loc[filtre, 'Revenue_USD'] = df.loc[filtre, 'Price_USD'] * df.loc[filtre, 'Units_Sold']
```

### Mettre dans un nouveau fichier qu'eyeliner de MAC

```python
df.to_csv('modified_mac.csv', index=False)
```
